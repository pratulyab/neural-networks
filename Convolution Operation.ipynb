```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('/Users/Pratulya/Documents/Wallpapers /197 Beautiful National Geographic Wallpapers 1600x1200/100.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# img = cv2.resize(img, (100,100))
img.shape
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray.shape
```

```python
plt.figure(0)
plt.imshow(img)
plt.axis("off")
plt.figure(1)
plt.imshow(gray, cmap='gray')
plt.axis("off")
plt.show()
```

```python
kernel = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])
kernel
```

```python
kernel = np.ones((3,3))/9
kernel
```

```python
convolved_img = np.zeros((gray.shape[0] - kernel.shape[0] + 1, gray.shape[1] - kernel.shape[1] + 1))
convolved_img.shape
```

```python
for i in range(convolved_img.shape[0]):
    for j in range(convolved_img.shape[1]):
        patch = gray[i:i+kernel.shape[0], j:j+kernel.shape[1]]
        h_prod = patch * kernel
        convolved_img[i,j] = max(0, h_prod.sum())
```

```python
plt.figure()
plt.imshow(convolved_img.astype('uint8'), cmap='gray')
plt.show()
```
