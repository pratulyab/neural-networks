```python
import numpy as np
import matplotlib.pyplot as plt

import keras
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
```

# Dropout

```python
x = np.linspace(-1, 1, 20)
y = x + 3*np.random.random((20,))
```

```python
x.shape, y.shape
```

```python
# test data
test_x = np.linspace(-1, 1, 20)
test_y = test_x + 3*np.random.random((20,))
```

```python
test_x.shape, test_y.shape
```

```python
plt.figure()
plt.scatter(x, y, c='magenta', s=50, alpha=0.5, label='train')
plt.scatter(test_x, test_y, c='cyan', s=50, alpha=0.5, label='test')
plt.legend(loc='upper left')
plt.ylim((-2.5, 2.5))
plt.show()
```

```python
overfitted = Sequential()
overfitted.add(Dense(300, input_shape=(1,)))
overfitted.add(Activation('relu'))
overfitted.add(Dense(300, activation='relu'))
overfitted.add(Dense(1))

overfitted.summary()
overfitted.compile(loss='mse', optimizer='rmsprop')
```

```python
constrained = Sequential()
constrained.add(Dense(300, input_shape=(1,)))
constrained.add(Dropout(0.4))
constrained.add(Activation('relu'))
constrained.add(Dense(300, activation='relu'))
constrained.add(Dense(1))

constrained.summary()
constrained.compile(loss='mse', optimizer='rmsprop')
```

```python
for e in range(500):
    overfitted.fit(x, y, shuffle=True, verbose=0)
    constrained.fit(x, y, shuffle=True, verbose=0)
    
    if e % 10 == 0:
        plt.figure()
        plt.scatter(x, y, c='magenta', s=50, alpha=0.5, label='train')
        plt.scatter(test_x, test_y, c='cyan', s=50, alpha=0.5, label='test')
        plt.plot(test_x, overfitted.predict(test_x), 'r-', lw=3, label='overfitting')
        plt.plot(test_x, constrained.predict(test_x), 'b--', lw=3, label='constrained')
        print('---- TESTING LOSS ----')
        print('Overfitting Loss=%.4f' % overfitted.evaluate(test_x, test_y, verbose=0))
        print('Constrained Loss=%.4f' % constrained.evaluate(test_x, test_y, verbose=0))
        print('----------------------')
        plt.legend(loc='upper left')
        plt.ylim((-2.5, 2.5))
        plt.show()
```

# Batch Normalization Intuition

```python
mean_01 = np.array([1.0, 4.0])
mean_02 = np.array([10.0, -4.0])

cov_01 = np.array([[3.0, 0.0],[0.0, 300.0]])
cov_02 = np.array([[3.0, 1.0],[1.0, 3.0]])

data_01 = np.random.multivariate_normal(mean_01, cov_01, 500)
data_02 = np.random.multivariate_normal(mean_02, cov_02, 500)
```

```python
plt.xlim(-20, 20)
plt.ylim(-80, 80)
plt.scatter(data_01[:, 0], data_01[:, 1], color='red')
plt.scatter(data_02[:, 0], data_02[:, 1])
plt.show()
```

```python
cov_03 = np.array([[5.0, 0.0],[0.0, 5.0]])
mean_03 = np.array([10.0, 40.0])
```

```python
X1 = (data_01 - data_01.mean(axis=0)) / np.std(data_01)
X2 = (data_02 - data_02.mean(axis=0)) / np.std(data_02)
print (X2.mean())
```

```python
# plt.xlim(-20, 20)
# plt.ylim(-80, 80)
plt.scatter(X1[:, 0], X1[:, 1], color='r')
plt.scatter(X2[:, 0], X2[:, 1])
plt.show()
```

```python
x1 = X1.dot(cov_03) + mean_03
x2 = X2.dot(cov_03) + mean_03

x1.shape, x2.shape
```

```python
plt.xlim(-20, 20)
plt.ylim(-80, 80)
plt.scatter(x1[:, 0], x1[:, 1], color='r')
plt.scatter(x2[:, 0], x2[:, 1])
plt.show()
```
